# FINAL PROJECT

```python
import pandas as pd 
import numpy as np 
```

```python
train = pd.read_csv('train.csv')
train
```

## summary data

```python
train.info()
```

## convert binary gaji

```python
convert_gaji = {'<=7jt':0,'>7jt':1}
```

```python
train['Gaji'] = train['Gaji'].replace(convert_gaji)
```

```python
train
```

```python
train[['Kelas Pekerja','Gaji']].groupby(['Kelas Pekerja'],as_index=False).mean()
```

```python
train[['Pekerjaan','Gaji']].groupby(['Pekerjaan'],as_index=False).mean()
```

```python
train['Kelas Pekerja'] = train['Kelas Pekerja'].replace('?',np.nan)
train['Pekerjaan'] = train['Pekerjaan'].replace('?',np.nan)
```

```python
train.dropna(how='any',inplace=True)
```

```python
train.info()
```

```python
for c in train.columns:
    print ("---- %s ---" % c)
    print (train[c].value_counts())
```

## mengubah kelamin

```python
kelamin = {'Laki2':0,'Perempuan':1}
```

```python
train['Jenis Kelamin'].replace(kelamin,inplace=True)
```

```python
train
```

## mengubah pendidikan

```python
train['Pendidikan'].value_counts()
```

```python
pendidikan = {
    '1st-4th':1,
    '5th-6th':2,
    'SD':3,
    '7th-8th':4,
    '9th':5,
    '10th':6,
    '11th':7,
    '12th':8,
    'SMA':9,
    'Sekolah Professional':10,
    'D3':11,
    'D4':12,
    'Pendidikan Tinggi':13,
    'Sarjana':14,
    'Master':15,
    'Doktor':16    
}
```

```python
train['Pendidikan'] = train['Pendidikan'].replace(pendidikan)
```

## get dummy variables

```python
train_baru = pd.get_dummies(train)
```

```python
train_baru
```

## melakukan train test

```python
from sklearn.model_selection import train_test_split
```

```python
X = train_baru.drop(['id','Gaji'],axis=1)
X
```

```python
y = train_baru['Gaji']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8)
```

## percobaan pake decission tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

model = DecisionTreeClassifier()
parameter = {'criterion':['gini','entropy'],'splitter':['best', 'random'],'max_depth':np.arange(1,10)}
gscv = GridSearchCV(model, param_grid=parameter, scoring='roc_auc',cv=5)
gscv.fit(X_train,y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
roc_auc_score(y_test,DecisionTreeClassifier(criterion='gini', splitter='best',max_depth=8).fit(X_train,y_train).predict(X_test))
```

## percobaan pake knn

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
parameter = {'n_neighbors':np.arange(1,100),'weights':['distance', 'uniform']}
gscv = RandomizedSearchCV(model, param_distributions=parameter, n_iter=50, scoring='roc_auc',cv=5)
gscv.fit(X_train,y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

## percobaan pake GBC

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model = GradientBoostingClassifier()
parameter = {'warm_start':[False,True]}
gscv = GridSearchCV(model, param_grid=parameter, scoring='roc_auc',cv=5)
gscv.fit(X_train,y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
hasil_predik = gscv.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
roc_auc_score(y_test,GradientBoostingClassifier(warm_start=True).fit(X_train,y_train).predict(X_test))
```

## percobaan pake random forest 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
parameter = {'warm_start':[False,True], 'bootstrap':[False,True]}
gscv = GridSearchCV(model, param_grid=parameter, scoring='roc_auc',cv=5)
gscv.fit(X_train,y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
roc_auc_score(y_test,RandomForestClassifier(bootstrap=True, oob_score=False, warm_start=True).fit(X_train,y_train).predict(X_test))
```

```python
# Code here
import pandas as pd 
import numpy as np 

gender = pd.read_csv('gender_submission.csv')
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
passid = test['PassengerId']
```

```python
y = train['Survived']
X = train.drop(['Name','Ticket','Fare','Embarked','Survived','PassengerId','Cabin'],axis=1)
sex = {'male':0,'female':1}
```

```python
X['Sex'] = X['Sex'].replace(sex)
```

```python
X['Age'] = X['Age'].fillna(X['Age'].mean())
```

```python
X = X.round(0)
```

```python
test
```

```python
X["Age"].isna().sum()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

rand = np.arange(1,500)
model = KNeighborsClassifier()
parameter = {'n_neighbors':rand, 'weights':['distance', 'uniform']}
```

```python
gscv = GridSearchCV(model, param_grid=parameter, scoring='accuracy', cv=5)
```

```python
gscv.fit(X,y)
```

```python
gscv.best_params_
```

```python
test
```

```python
test = test.drop(['Name','Ticket','Fare','Embarked','PassengerId','Cabin'],axis=1)
```

```python
test['Sex'] = test['Sex'].replace(sex)
test['Age'] = test['Age'].fillna(test['Age'].mean())
```

```python
test
```

```python
hasil = KNeighborsClassifier(n_neighbors= 12, weights='distance').fit(X,y).predict(test)
```

```python
submitan = pd.DataFrame([passid,hasil]).T
submitan
```

```python
submitan.columns = ['PassengerId','Survived']
```

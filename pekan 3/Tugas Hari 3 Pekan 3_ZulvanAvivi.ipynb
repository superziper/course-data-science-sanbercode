```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(20,10))
ax[0].plot(x, tan, color='b', label='tanX')
ax[0].set_title('Grafik tanx')
ax[1].plot(x, sin, color='r', label='sinX')
ax[1].set_title('Grafik sinx')
ax[2].plot(x, cos, color='g', label='cosX')
ax[2].set_title('Grafik cosx')
fig.legend(loc='upper center')
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y1 = [(np.cos(x**2)),(np.cos(x**3)),(np.cos(x**4)),(np.cos(x**5))]
color = ['r','g','k','y']
```

```python
# code here
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(x, y)
ax.set_title('Grafik CosX')

axis = [ax.inset_axes([-0.1,-0.1,0.3,0.3]),ax.inset_axes([0.9,-0.1,0.3,0.3]),ax.inset_axes([0.9,0.9,0.3,0.3]), ax.inset_axes([-0.1,0.9,0.3,0.3])]

for i in range(4):
    axis[i].plot(x,y,color='b')
    axis[i].plot(x,y1[i],color=color[i])
    axis[i].set_title('Grafik CosX^'+str(i+2))
```
